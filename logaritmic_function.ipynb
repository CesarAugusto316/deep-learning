```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0.001, 1, 200)

logx = np.log(x)
log10x = np.log10(x)
plt.figure(figsize=(12, 6))
plt.rcParams.update({'font.size': 16})
plt.plot(x, logx, label=r'$\log(x)$')
plt.plot(x, log10x, label=r'$\log_{10}(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
plt.show( )
```
